```
import requests
import pandas as pd
```

```
url_info_cadastro = 'http://dados.cvm.gov.br/dados/FI/CAD/DADOS/inf_cadastral_fi_20200124.csv'
df_cadastro = pd.read_csv(url_info_cadastro, sep=';', encoding='latin1')
df_cadastro.info()
```

```
urls_eventual_fi = [
    'http://dados.cvm.gov.br/dados/FI/DOC/EVENTUAL/DADOS/eventual_fi_2015.csv',
    'http://dados.cvm.gov.br/dados/FI/DOC/EVENTUAL/DADOS/eventual_fi_2016.csv',
    'http://dados.cvm.gov.br/dados/FI/DOC/EVENTUAL/DADOS/eventual_fi_2017.csv',
    'http://dados.cvm.gov.br/dados/FI/DOC/EVENTUAL/DADOS/eventual_fi_2018.csv',
    'http://dados.cvm.gov.br/dados/FI/DOC/EVENTUAL/DADOS/eventual_fi_2019.csv',
    'http://dados.cvm.gov.br/dados/FI/DOC/EVENTUAL/DADOS/eventual_fi_2020.csv'
]

df_info_eventual =  pd.concat([
    pd.read_csv(urls_eventual_fi[0], sep=';', encoding='latin1', error_bad_lines=False, quoting=3),
    pd.read_csv(urls_eventual_fi[1], sep=';', encoding='latin1', error_bad_lines=False, quoting=3),
    pd.read_csv(urls_eventual_fi[2], sep=';', encoding='latin1', error_bad_lines=False, quoting=3),
    pd.read_csv(urls_eventual_fi[3], sep=';', encoding='latin1', error_bad_lines=False, quoting=3),
    pd.read_csv(urls_eventual_fi[4], sep=';', encoding='latin1', error_bad_lines=False, quoting=3)
])

df_info_eventual.info()
```

```
# sort by admin
df_cadastro.sort_values(by='ADMIN')
df_cadastro.head()
```

```
# faz um select distinct nos admins
df_cadastro['ADMIN'].unique()
```

```
# filtra apenas os fundos administrados pela CAIXA
df = df_cadastro[df_cadastro['ADMIN'] == 'CAIXA ECONOMICA FEDERAL']

# filtra apenas os fundos ativos
df = df[df['SIT'] == 'EM FUNCIONAMENTO NORMAL']

# formata o campo CO_PRD com o cnpj 
df['CO_PRD'] = df['CNPJ_FUNDO'].str.replace(".", "")
df['CO_PRD'] = df['CO_PRD'].str.replace("-", "")
df['CO_PRD'] = df['CO_PRD'].str.replace("/", "")

# remove as colunas que não serão utilizadas
df.drop('DT_REG', axis=1, inplace=True)
df.drop('DT_CANCEL', axis=1, inplace=True)
df.drop('SIT', axis=1, inplace=True)
df.drop('DT_INI_SIT', axis=1, inplace=True)
df.drop('DT_INI_EXERC', axis=1, inplace=True)
df.drop('DT_FIM_EXERC', axis=1, inplace=True)
df.drop('DIRETOR', axis=1, inplace=True)
df.drop('CNPJ_ADMIN', axis=1, inplace=True)
df.drop('ADMIN', axis=1, inplace=True)
df.drop('PF_PJ_GESTOR', axis=1, inplace=True)
df.drop('CNPJ_AUDITOR', axis=1, inplace=True)
df.drop('AUDITOR', axis=1, inplace=True)
df.drop('DT_CONST', axis=1, inplace=True)
df.drop('RENTAB_FUNDO', axis=1, inplace=True)
df.drop('CONDOM', axis=1, inplace=True)
df.drop('TAXA_PERFM', axis=1, inplace=True)
df.drop('VL_PATRIM_LIQ', axis=1, inplace=True)
df.drop('DT_PATRIM_LIQ', axis=1, inplace=True)
df.drop('CPF_CNPJ_GESTOR', axis=1, inplace=True)

print(df.head())

df.to_excel('fundos_admin_caixa.xlsx')
```

```
# formata o campo CO_PRD com o cnpj 
df_info_eventual['CO_PRD'] = df_info_eventual['CNPJ_FUNDO'].str.replace(".", "")
df_info_eventual['CO_PRD'] = df_info_eventual['CO_PRD'].str.replace("-", "")
df_info_eventual['CO_PRD'] = df_info_eventual['CO_PRD'].str.replace("/", "")

df_info_eventual.head()

df_inner = pd.merge(left=df,right=df_info_eventual,left_on='CO_PRD',right_on='CO_PRD')

df_inner.drop('CNPJ_FUNDO_y', axis=1, inplace=True)
df_inner.drop('DENOM_SOCIAL_y', axis=1, inplace=True)
df_inner.drop('NM_ARQ', axis=1, inplace=True)

df_inner.sort_values(by=['CO_PRD', 'DT_COMPTC'], inplace=True)

df_inner.to_excel('fundos_admin_caixa_documentos.xlsx')
```

```
df_regulamento = df_inner[df_inner['TP_DOC'] == 'REGUL FDO']

# pega a última alteração de regulamento por fundo
df_regulamento = df_regulamento.sort_values('DT_RECEB').groupby('CO_PRD').tail(1)

df_regulamento.head()

df_regulamento.to_excel('fundos_admin_caixa_ultimos_regulamentos.xlsx')
```
